**1757. Recyclable and Low Fat Products**

import pandas as pd

def find_products(products: pd.DataFrame) -> pd.DataFrame:

**Intuition**

When I first read this problem, my intuition was to use a simple filtering approach to find the products that meet both conditions: low fat and recyclable. I thought about using a boolean mask to select only the rows that satisfy both conditions.

**Approach**

My approach is to create a boolean mask mask that selects only the rows where both low_fats and recyclable are 'Y'. Then, I use this mask to filter the original dataframe products and select only the column product_id.

Here's a step-by-step breakdown:


    Create a boolean mask mask using the conditions (products['low_fats']== 'Y') & (products['recyclable'] == 'Y').

    Use this mask to filter the original dataframe products and select only the rows where both conditions are true.

    From the filtered dataframe, select only the column product_id.

**Time Complexity** : O(N)

**Space Complexity**  : O(1)




```python
import pandas as pd

def find_products(products: pd.DataFrame) -> pd.DataFrame:
    mask = (products['low_fats']== 'Y') & (products['recyclable'] == 'Y')
    return products[mask][['product_id']]
```
